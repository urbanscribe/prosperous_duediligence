```python
df = pd.read_csv("SBCscreen.csv")
import pandas as pd

def convert_column(df, col_name):
    df[col_name] = df[col_name].apply(lambda x: convert_to_float(x))
    print("done with:", col_name)

def convert_to_float(x):
    if type(x) == str:
        if x[-1] == 'M':
            return float(x[:-1]) * 1e6
        elif x[-1] == 'B':
            return float(x[:-1]) * 1e9
        elif x[-1] == 'T':
            return float(x[:-1]) * 1e12
        elif x[-1] == 'k':
            return float(x[:-1]) * 1e3
    try:
        return float(x)
    except ValueError:
        return float('nan')


cols_to_convert = ['SBC', 'Market Cap', 'ValTraded', 'Market Cap.1', 'PriceD-1']
for col in cols_to_convert:
    convert_column(df, col)
# df.to_csv("converted_file.csv", index=False, float_format="%.2f")

df = df[df['SBC'].notnull()]
df
```

```python
df['SBC/MktCap'] = df['SBC'] / df['Market Cap']
df
```

```python
df.drop('PriceD-1', axis=1, inplace=True)
df.drop('ValTraded', axis=1, inplace=True)
df.drop('Market Cap.1', axis=1, inplace=True)
df
```

```python
df['Market Cap']=df['Market Cap'].apply(lambda x: "{:,.0f}".format(x))
df['SBC']=df['SBC'].apply(lambda x: "{:,.0f}".format(x))
df['SBC/MktCap']=df['SBC/MktCap'].apply(lambda x: "{:.1%}".format(x))
df
```

```python
df.sort_values("SBC/MktCap", ascending=False, inplace=True)
df
```

```python
df.to_csv("sortedfile.csv")
```
